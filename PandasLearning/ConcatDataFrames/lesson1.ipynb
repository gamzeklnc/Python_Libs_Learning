```python

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset("lukebarousse/data_jobs")
df=dataset['train'].to_pandas()

df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
```

```python
df["job_posted_month"]=df["job_posted_date"].dt.strftime('%b')#ayın ilk 3 harfi
df
```

```python
"""dict_months={
"Jan":df_jan,
"Feb":df_feb,
"Mar":df_mar, 
"Apr":df_apr,
"May":df_may,
"Jun":df_jun,
"Jul":df_jul,
"Aug":df_aug,
"Sep":df_sep,
"Oct":df_oct,
"Nov":df_nov,
"Dec":df_dec


}"""
#{KEY: VALUE for ITEM in LİSTE}

months = df["job_posted_month"].unique()
months
dict_months={month:df[df["job_posted_month"]==month] for month in months}
dict_months["Jan"]
```

BOSS NEEDS 2023 QUARTER 1

```python
df_q1=pd.concat([dict_months["Jan"],dict_months["Feb"],dict_months["Mar"]] , ignore_index=True)
#concat satır ve sütun birleştirmede kullanılır
```

```python
df_q1["job_posted_month"].value_counts().plot(kind="bar", color="red")
plt.xticks(rotation=0)
plt.show()
```

```python
#pip install datasets 
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset("lukebarousse/data_jobs")
df=dataset['train'].to_pandas()

df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
df["job_skills"]=df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
df["job_work_from_home"].value_counts().plot(kind='pie',startangle=180, autopct='%1.1f%%', colors=["#2db190","#B159CC"])
plt.title("Work From Home Status")
plt.show

#startangle=0 (veya hiç yazmazsan): Dilimler sağdan başlar.
#startangle=90:Dilimler alttan başlar.
#startangle=180:Dilimler soldan başlar (senin örneğinde olduğu gibi).
#startangle=270:Dilimler üstten başlar.
```

```python
df["job_title_short"].value_counts().plot(kind='pie')
plt.title("Job Title in Luke's Dataset ")
plt.ylabel("")
plt.show()
```

```python
df["job_title_short"].value_counts().plot(kind='barh')
plt.title("Job Title in Luke's Dataset ")
plt.gca().invert_yaxis()  # Y eksenini ters çevirir
plt.ylabel("")
plt.show()
```

```python
fig ,ax= plt.subplots() #figür boyutunu ayarlar.

ax.pie(df["job_work_from_home"].value_counts())
plt.show()
```

```python
fig ,ax= plt.subplots(1,3) #figür boyutunu ayarlar.
dict_column={
    "job_work_from_home":"Work From Home",
    "job_no_degree_mention":"Job Degree Req.",
    "job_health_insurance":"Health Insurance Offered"
}
for i,(column,title) in enumerate(dict_column.items ()):
    # print(i,column,title)
    ax[i].pie(df[column].value_counts(),startangle=90, autopct='%1.1f%%', colors=["#2db190","#B159CC"],labels=[False,True])
    ax[i].set_title(title)
plt.show()
```

```python
#pip install datasets 
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset("lukebarousse/data_jobs")
df=dataset['train'].to_pandas()

df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
```

```python
job_counts=df.job_title_short.value_counts()
```

```python
job_counts.plot(kind='barh')
```

```python
job_counts.plot(kind='line')
```

```python
job_counts.plot(kind='bar')
```

```python
plt.bar(job_counts.index,job_counts)
plt.title("Posting by Job Title")
plt.ylabel("Number of Postings")
plt.xlabel("")
plt.xticks(rotation=60, ha='right')# rotate x-axis labels for better readability(boluk vererek)
plt.show()
```
